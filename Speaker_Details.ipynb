```python
from bs4 import BeautifulSoup
import requests
url = "https://www.databricks.com/dataaisummit/speakers?itm_data=hp-featured-speakers-dais23"  # Replace with the URL of the website you want to scrape
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
topic_selection_class="css-1oupgqn"
```

```python
all_p_tags = soup.find_all('p',{'class': topic_selection_class})
```

```python
print(soup.find(topic_selection_class))
```

```python
speakers=soup.find_all("div","css-1oupgqn" )
speaker_name= "css-78idzx"
speaker_job_title= "Speaker-jobTitle css-1c6h1l2"
speaker_company= "css-b9oelt"
```

```python
import pandas as pd
df = pd.DataFrame(columns=['Name', 'Role', 'Company'])
```

```python
for i in range(len(speakers)):

  name=speakers[i].find('p',speaker_name).text
  title=speakers[i].find('p',speaker_job_title).text
  company=speakers[i].find('p',speaker_company).text

  row_data = {'Name': name, 'Role': title, 'Company': company}

    # Append the dictionary as a new row in the DataFrame
  df = df.append(row_data, ignore_index=True)
```

```python
df.to_csv('Details_of_the_speakers.csv', index=False)
```
